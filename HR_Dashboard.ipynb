# CREATING CHARTS

```python
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Employee Count By Job Title

no_of_emp_by_dept=HR_df['Job Title'].value_counts()
no_of_emp_by_dept
```

```python
plt.title('Employee Count By Job Title')

plt.barh(no_of_emp_by_dept.index,no_of_emp_by_dept,color='orange');
```

```python
#GENDER DISTRIBUTION OF THE STAFF
HR_df.Gender.unique()
Gender_counts=HR_df.Gender.value_counts()
Gender_counts
```

```python
plt.title('Employee Count By Job Title')
#plt.pie(Gender_counts);
#plt.legend(['female','male'])
#plt.pie(Gender_counts,labels=Gender_counts.index,autopct='%.1f%%');
plt.pie(Gender_counts,labels=Gender_counts.index,autopct='%.1f%%',startangle=90);
```

```python
#AGE SPREAD OF THE STAFF
plt.title('AGE SPREAD OF THE STAFF')
plt.hist(HR_df.Age,bins=np.arange(10,80,10))
plt.xlabel('Age')
```

```python
#T0P EARNING EMPLOYEES
HR_df.head(10)
```

```python
#top_earners= HR_df.sort_values('Salary',ascending=False).head(5)


top_earners= HR_df.sort_values('Salary',ascending=True).tail(5)
top_earners
top_earners[::-1]
```

```python
plt.title('Top earning Employees')
plt.barh(top_earners.Name,top_earners['Salary'],color='Orange',align='center');
plt.xlabel('Salary');
#plt.xticks([])
```

```python
#HEADCOUNT TREND OVER TIME
HR_df
```

```python
#Extracting year from date of join column
HR_df['year']=pd.DatetimeIndex(HR_df['Date of Join']).year
```

```python
HR_df
```

```python
#Employee count over years
employee_per_year = HR_df.groupby('year').size()
employee_per_year
```

```python
plt.figure(figsize=(10, 5))
plt.plot(employee_per_year.index,employee_per_year.values, marker='o', linestyle='-', color='#20283E');
```

# Salary Based on Education

```python
import seaborn as sns
sns.boxplot(data=HR_df,x='Education Qualification', y='Salary')
```

# DASHBOARD

```python
import matplotlib.pyplot as plt



fig,axes=plt.subplots(2,3,figsize=(18, 10))
plt.suptitle("HR Data Dashboard â€“ Employee Insights", fontsize=24, fontweight='bold')
plt.tight_layout(pad=4)

#Top earning Employees
axes[0,0].set_title('Top earning Employees')
colors=["#9467BD", "#2CA02C", "#1F77B4", "#FF7F0E","#D62728"]
axes[0,0].barh(top_earners.Name,top_earners['Salary'],color=colors, align='center');
axes[0,0].set_xlabel('Salary');

#Employee count over years
axes[0,1].set_title('Employee Count Over the Years')
axes[0,1].plot(employee_per_year.index,employee_per_year.values, marker='o', linestyle='-', color='#20283E') 
#axes[0,1].set_xlabel('Year')
axes[0,1].set_ylabel('Employee count')


#Gender Distribution of the staff
axes[0,2].set_title('Gender Distribution of the Staff')
axes[0,2].pie(Gender_counts,labels=Gender_counts.index,autopct='%.1f%%',startangle=90);

#Employee Count By Job Title
axes[1,0].set_title('Employee Count By Job Title')
axes[1,0].barh(no_of_emp_by_dept.index,no_of_emp_by_dept,color="#1F77B4");

#Age Spread of the Staff
axes[1,1].set_title('AGE SPREAD OF THE STAFF')
axes[1,1].hist(HR_df.Age,bins=np.arange(10,80,10),color='#FF7F0E');
axes[1,1].set_ylabel('No of staffs');

#Salary Based On Education
axes[1,2].set_title('Salary Based On Education')
sns.boxplot(data=HR_df,x='Education Qualification', y='Salary',ax=axes[1,2]);


```
